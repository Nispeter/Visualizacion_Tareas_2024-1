```python
# ordenamos el dataframe por el total de inmigrantes en el periodo, en orden descendente
df_chi.sort_values(by='Total', ascending=False, axis=0, inplace=True)
# obtenemos los 5 países con mayor valor total 
df_top5 = df_chi.head(5)
# realizamos la traspuesta del dataframe
df_top5 = df_top5[years].transpose() 

df_top5.index = df_top5.index.map(int) 

# cambiamos los índices a entero, para favorecer el ploteo
df_top5.plot(kind='line', figsize=(20,8)) 

plt.xticks(np.arange(min(df_top5.index), max(df_top5.index)+1, 1.0))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) 
df_top5.plot(kind='area', #sólo cambiamos el valor del parámetro kind para obtener un area chart
             stacked=False,
             figsize=(20, 8), 
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.xticks(np.arange(min(df_top5.index), max(df_top5.index)+1, 1.0))

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) 
df_top5.plot(kind='area',
             alpha=0.25, # variamos el valor de transparencia (0-1, por defecto 0.5)
             stacked=False,
             figsize=(20, 8), 
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.xticks(np.arange(min(df_top5.index), max(df_top5.index)+1, 1.0))

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) 
df_top5.plot(kind='area',
             alpha=1, # sin transparencia
             stacked=True, # área apilada
             figsize=(20, 8), 
             )

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.xticks(np.arange(min(df_top5.index), max(df_top5.index)+1, 1.0))

plt.show()
```

```python
# analicemos la inmigración desde los países escandinavos
df_chi.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_chi_escand = df_chi.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_chi_escand.head()
```

```python
# generamos histograma
df_chi_escand.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 2005 - 2016')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# obtenemos los ticks del eje x de acuerdo a los valores de los bins
count, bin_edges = np.histogram(df_chi_escand, 10)

# describimos un histograma desapilado (por defecto)
df_chi_escand.plot(kind ='hist', 
          figsize=(10, 6),
          bins=10,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 2005 - 2016')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_chi_escand, 10)
xmin = bin_edges[0] - 10   #  el primer valor de bin es 26.0, se añade un buffer de 10 por motivos de presentación
xmax = bin_edges[-1] + 10  #  el último valor de bin es 111.0, se añade un buffer de 10 por motivos de presentación

# histograma apilado
df_chi_escand.plot(kind='hist',
          figsize=(10, 6), 
          bins=10,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 2005 - 2016')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```
